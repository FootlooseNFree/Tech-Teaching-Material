````python
#@title Import pandas and sqlite3 libraries
import pandas as pd
import sqlite3
import time
print("\033[92mLibraries imported successfully\033[00m")
````

````python
#@title Import the data
url = "https://raw.githubusercontent.com/FootlooseNFree/my_files/refs/heads/main/netflix_data.csv"
start_time = time.perf_counter()
df = pd.read_csv(url)
end_time = time.perf_counter()
df.dropna(inplace=True)
print('\033[92mRows with missing values were dropped\033[00m')
df['imdbNumVotes'] = df['imdbNumVotes'].astype(int)
print(f"\033[92mData imported successfully\033[00m")
print(f"Time taken to import the data: \033[92m{end_time - start_time:0.4f} seconds\033[00m")
````

````python
#@title Initiate the SQL environment
# Define a function for reuse
def run_query():
  con = sqlite3.connect('Netflix.db')
  cur = con.cursor()

  # Create Countries table and add the data to it
  df.to_sql('Netflix', con, if_exists='replace', index=False)
  pd.set_option('display.max_columns', None)
  pd.set_option('display.max_rows', None)

  # Print the table name
  print('Table name: \033[91mNetflix\033[00m')

  # Print the column names
  print('Column names: \033[92m' + str(list(df.columns)) + '\033[00m')
  print()

  query = input('Enter your SQL Query:  ')
  try:
    df_output = pd.read_sql_query(query, con)
  except Exception as e:
    print(e)
    print('Try again!!!')
    return None

  con.close()
  return df_output
````

For your convenience, running each code block below the respective question will also print the table name in red, and column names in green.

To get acquainted with your data, run a query to find the total number of `title`s.

````python
query_result = run_query()
query_result
````
